This notebook is just graphing our results that we found so the trends are more identifiable.

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ensemble Methods Data
x1 = ["Median", "Mean", "Weighted Average"]
y1 = [0.36502, 0.37424, 0.37440]

x2 = ["Median", "Mean", "Weighted Average"]
y2 = [0.36594, 0.37464, 0.37687]
```

```python
# Plotting the lines
plt.plot(x1, y1, label='3 Models', marker='o')
plt.plot(x2, y2, label='4 Models', marker='o')

# Adding labels and title
plt.xlabel('Method')
plt.ylabel('MAP Score')
plt.title('Ensemble Methods Comparison')
plt.legend()

# Show ticks
# plt.grid()

# Displaying the plot
plt.show()
```
